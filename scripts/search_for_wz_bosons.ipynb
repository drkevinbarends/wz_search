#### Bosons

The SM describes the electromagnetic force, the weak nuclear force, and the strong nuclear force, but is yet to describe the gravitational force. Bosons are fundamental particles that either have spin-1 or spin-0. They mediate the interactions between particles and act as force carriers. #### Fermions

Fermions are fundamental particles with spin-1/2 and consist of leptons and quarks. Leptons consist of "leptons" and neutrinos, where the "leptons" are electrons, muons, and taus. Each "lepton" has an electrically neutral counterpart, called neutrinos, namely, electron neutrino, muon neutrino, and tau neutrino.  Quarks consist of six flavours, namely, up, down, charm, strange, bottom (or beauty), and top. There are three up-type quarks, namely, up, charm, and top, and they are so named because they have positive electric charge. There are also three down-type quarks, namely, down, strange, and bottom, because they have negative electric charge. Quarks have an additional charge called "colour", and there are three types of colour charge, namely, blue, green, and red. 

There are three generations of fermions, where the first generation is responsible for all matter you see around you, and each suuccessive generation of particles is heavier than the previous generations.

The strong nuclear force says that quarks cannot exist in isolation, i.e., they always exist in pairs. However, the strong nuclear force also says that quarks appear to be free at high enough energies, known as asymptotic freedom. Additionally, the strong nuclear force says that free quarks and gluons combine to form composite particles called hadrons, also called hadronization. This hadronization time occurs within 10^-23 seconds. 

All the fundamental particles in the SM have an anti-particle counterpart which is identical in every way except that they posses the opposite electrical charge.

#### Particle Interaction

According to the SM, there are only few types of interactions that are allowed. All electrically charged particles can interact via the photon. Since the photon itself is electrically neutral, it does not interact with itself. All colour charged particles can interact via the gluon. Since the gluon itself is colour charged, it can interact with itself. All fermions possess the weak charge and therefore can interact via the W or Z bosons. Since the W boson carries the weak charge, it can interact with itself. The Z boson does not carry the weak charge, so it cannot interact with itself. 

The photon and Z boson has no structural impact on the particle, e.g., an electron to an electron. The gluon has no structural impact on the particle, but does change its colour charge, e.g., green top quark to anti-blue top quark. The W boson is the only boson interaction that changes the flavour of the particle. The W boson changes an up-type quark into a down-type quark, or vice versa, e.g., an top quark into a bottom quark. Additionally, the W boson changes a lepton into its corresponding lepton-neutrino e.g., electron to electron neutrino.

#### Particle Decay

All unstable particles will decay. All non-fundamental particles will split into their internal constituents when they decay. However, all fundamental particles have no internal constituents, so when they decay, they will transform all their energy and mass into new lighter particles. The allowed SM interactions also tell us how particles decay. Since the photon, gluon, and Z boson do not change the flavour of the particle - do not transform into new particles - they cannot be used to represent decays. Therefore, the W boson represents particle decays. However, due to the fact that the Z boson has mass, it can also decay.

Additionally, particles can annihilate to form other particles. A fermion interacting with its corresponding anti-fermion, will produce a boson. Furthermore,photons and gluons can undergo pair production, where the photon produces a lepton and its corresponding anti-lepton, while the gluon produces a quark and its corresponding anti-quark.

#### WZ Importance

The W and Z bosons are the fundamental particles responsible for mediating the weak nuclear force, which is the force behind radioactive decay and the nuclear fusion that powers the Sun. While the top quark is the heaviest particle of matter, the W boson ($m_W \approx 80\text{ GeV}$) and Z boson ($m_Z \approx 91\text{ GeV}$) are among the heaviest carriers of forces in the Standard Model (SM). Because they are so massive, they are deeply connected to the Higgs mechanism, which is the process that gives particles their mass. Studying how W and Z bosons are produced together allows us to directly test our understanding of how forces and mass interact at the most fundamental level.

A unique feature of W and Z bosons is that, unlike the carriers of electromagnetism (photons), they carry weak charges themselves. This means that W and Z bosons can actually interact, collide, and scatter off one another. When we search for the simultaneous production of a W and a Z boson (called WZ diboson production), we are looking at a process where force carriers directly interact to create more force carriers. By measuring the rate and angles at which these WZ pairs are produced, we can perform precision measurements on these interactions. Any subtle deviation from the SM predictions would instantly signal that our mathematical understanding of forces is incomplete.

Furthermore, because WZ production has a very distinct experimental signature—decaying cleanly into three high-energy leptons (like electrons or muons) with minimal background noise—it serves as an ideal environment to search for new physics. WZ boson pairs are strongly linked to various theories that go Beyond the Standard Model (BSM). For example, many models predict the existence of ultra-heavy, undiscovered particles that would preferentially decay directly into a W and a Z boson.

These protons are bunched up together before colliding at the centre of the four major experiments. The aftermath of these collisions are then captured by the detectors and the ttbar is stored for further analysis.

### The ATLAS Detector

The ATLAS detector is the largest detector in the world and is cylindrical in shape, which is about 45 metres long and 25 metres in diameter. Many subdetectors make the ATLAS detector starting from the Inner Detector, which is the closest subdetector to the collision point, all the way to the Muon Spectrometer, which is the outermost subdetector. Additionally, large magnets are used to focus the proton bunches and to help identify particles that come from the collision.

The Inner Detector (ID) is used to measure the trajectory of the particles that come from the collisions. These trajectories are reconstructed from the signals that are left in the detector as the particles interact with the material of the sensors. The next subdetector is the Electric Calorimeter (ECal) which is used to measure the energy of electromagnetically particles, specifically, electrons and photons. The Hadronic Calorimeter (HCal) follows and is used to measure the energy of hadrons. The last subdetector is the Muon Spectrometer (MS) which is used to measure the tracks and energy of muons.

### Experimental Signature
 
Once a W boson is produced, it can decay hadronically or leptonically. Once a Z boson is produced, is can decay into opposite-sign same flavour quarks, leptons, or neutrinos. Together, there numerous decay modes that can happen and therefore, we need to pick which decay mode to perform the search.

Let's say we pick the decay mode with the highest decay ratio, i.e., the W and Z bosons decayed hadronically, as a W boson decays hadronically about 67% of the time and a Z boson decays hadronically 70% of the time. This means our signature should be 

$$WZ \rightarrow (q_{1}q_{2})(q_{3}\bar{q_{3}})$$

This means we are looking for a signature of at least 4 jets (i.e., 2+2). This signature has both theoretical and experimental advantages and disadvantages. Due to these disadvantages, we instead perform the search where both the W and Z bosons decay leptonically, i.e.

$$WZ \rightarrow (l_{1}\nu_{1})(l_{2}\bar{l_{2}})$$

#### Number of events

The number of events observed in an experiment depends on two factors:
1. How frequently the process occurs.
2. How much data have been collected.

The amount of data collected is quantified using the **integrated luminosity**. A larger integrated luminosity corresponds to more proton-proton collisions and therefore more opportunities for rare processes to occur.

Therefore, to determine the number of simulation events we should theoretically produce, we use 

$$N = \mathcal{L}\sigma$$

where $N$ is the number of events, $\mathcal{L}$ is the integrated luminosity, and $\sigma$ is the cross-section of that specific process.

##### Event weights

Imagine that two simulation samples each contain 100,000 events. Does this mean the corresponding physical processes occur equally often in nature? Not necessarily. One process may occur much more frequently than the other. The number of simulated events is chosen by physicists and does not directly reflect the true production rate of the process. Event weights allow us to convert simulation samples into realistic predictions.

In a full particle physics analysis, event weights include contributions from:

- production values (cross-sections, k-factors, filter efficiencies)
- generation weights
- Detector efficiencies
- Trigger efficiencies
- Reconstruction efficiencies
- Additional corrections

Each simulated event contributes a certain amount to the final prediction. This contribution is quantified using an event weight. The formula we'll be using is as follow:

$$\text{event weight} = \mathcal{L}\sigma\prod\epsilon_{ff}$$

These factors allow simulation to be converted into a prediction for the number of events expected in the detector.

```python
lumi = 36646.7 

def calc_weight(data):
    scale_factors = (data["ScaleFactor_PILEUP"] * data["ScaleFactor_ELE"] * data["ScaleFactor_MUON"] * data["ScaleFactor_MLTRIGGER"])
    norm_factor = (data["xsec"] * data["kfac"] * data["filteff"]) / data["sum_of_weights"]
    weight = data["mcWeight"] * scale_factors * norm_factor * lumi
    return weight
```

```python
wz['eventWeight'] = calc_weight(wz)
diboson['eventWeight'] = calc_weight(diboson)
vjets['eventWeight'] = calc_weight(vjets)
topquarks['eventWeight'] = calc_weight(topquarks)
```

We can use our understanding of the physics to help us develop a trigger. This trigger will help us choose the "right" leptons. In this analysis, we are only looking for at least two muons and electrons, so we keep events where two muons or electrons fired this lepton trigger. 

```python
def cut_trig(trigDE, trigDM):
    return trigDE | trigDM

data = data[cut_trig(data.trigDE,data.trigDM)]
wz = wz[cut_trig(wz.trigDE,wz.trigDM)]
diboson = diboson[cut_trig(diboson.trigDE,diboson.trigDM)]
vjets = vjets[cut_trig(vjets.trigDE,vjets.trigDM)]
topquarks = topquarks[cut_trig(topquarks.trigDE,topquarks.trigDM)]
```

Now that we know what events fired the trigger, we now want to ensure we have the leptons in every event that fired that trigger. This is matching the lepton to the trigger. Since we have three leptons in every event, we want to have at least two of those leptons matched to the trigger.

```python
def cut_match_MLtrig(trigML):
    return trigML==1

data = data[cut_match_MLtrig(data.TriggerMatch_DILEPTON)]
wz = wz[cut_match_MLtrig(wz.TriggerMatch_DILEPTON)]
diboson = diboson[cut_match_MLtrig(diboson.TriggerMatch_DILEPTON)]
vjets = vjets[cut_match_MLtrig(vjets.TriggerMatch_DILEPTON)]
topquarks = topquarks[cut_match_MLtrig(topquarks.TriggerMatch_DILEPTON)]
```

Additionally, we know that the W boson also decays into a neutrino, and therefore we can apply a selection on the neutrinos. However, what selection? We know neutrinos are not detected and therefore go "missing". In ATLAS, there is a variable called Missing Transverse Energy or MET, that is used to represent all particles that are not detected within the ATLAS detector. This is done by adding up the square of the momenta of all the objects that are detected and taking the negative of that sum, since momentum must be conserved in the collision. In our signature, this MET variable will be used as a proxy for the neutrino from the W boson.

Our event signature has two parts:

We have three leptons and MET that accounts for our final state objects. We can re-engineer the parent objects they came from. For example, we can see if two out of the three leptons originated from a Z boson. How do we do that?

From relativistic kinematics, we know that we can combine two objects four momenta to determine the mass from what they originated. Since we want two leptons that came from the Z boson, we can see which two of the three came from the Z. We can't just cycle through the three because there are some extra details that we know from theory. We know that when a Z boson decays leptonically, it decays into opposite-sign same flavour (OSSF) leptons. This means we are looking for either two electrons or two muons that have opposite-sign. Additionally, if they originate from the Z boson, they will have a rest mass of around 91 GeV.
